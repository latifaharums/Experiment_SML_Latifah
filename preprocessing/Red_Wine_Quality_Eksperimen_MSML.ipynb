# **1. Perkenalan Dataset**


 **Sumber Dataset**:  
   Dataset diperoleh dari Kaggle ( https://www.kaggle.com/datasets/uciml/red-wine-quality-cortez-et-al-2009 )

**GOALS**:  Membangun model yang dapat memprediksi skor kualitas pada anggur merah berdasarkan karakteristik atau atribut dari suatu bahan (dalam hal ini, anggur merah) yang dapat diukur dan dijelaskan menggunakan prinsip-prinsip fisika dan kimia.

**POV:** Bisa digunakan untuk penelitian yang menghubungkan tes laboratorium dengan persepsi kualitas anggur, sehingga dapat membantu dalam kontrol kualitas dan sertifikasi.


> **Dataset "Red Wine Quality"** yaitu data dari uji fisikokimia dan penilaian sensori (kualitas) anggur merah dari varian "Vinho Verde" yang berasal dari wilayah utara Portugal. Setiap baris dalam dataset ini merepresentasikan satu sampel anggur merah, dan kolom-kolomnya mencatat berbagai hasil pengukuran laboratorium (fitur fisikokimia) serta skor kualitas yang diberikan oleh panelis ahli.

Atribut (kolom) yang ada pada dataset mencakup:

1.  **fixed acidity** (Keasaman Tetap): Jumlah asam-asam organik dalam anggur yang tidak mudah menguap saat anggur dipanaskan.
2.  **citric acid** (Asam Sitrat):  Jumlah asam sitrat dalam anggur. Asam sitrat sering ditambahkan ke anggur untuk meningkatkan keasaman total, menambah kesegaran, dan memberikan rasa buah-buahan sitrus.
3.  **residual sugar** (Sisa Gula): Jumlah gula (terutama fruktosa dan glukosa) yang tersisa dalam anggur setelah proses fermentasi selesai.
4.** volatile acidity** (Keasaman Mudah Menguap): Mengukur jumlah asam asetat dalam anggur. Asam asetat adalah asam yang mudah menguap.  
5. **chlorides **(Klorida): Konsentrasi garam klorida dalam anggur. Tingkat klorida dapat dipengaruhi oleh praktik vitikultur (budidaya anggur) dan jenis tanah.
6. **free sulfur dioxide** (Sulfur Dioksida Bebas): Bagian dari total sulfur dioksida yang ada dalam bentuk bebas (tidak terikat dengan senyawa lain) di dalam anggur. SO2 bebas berfungsi sebagai antioksidan dan antimikroba, melindungi anggur dari pembusukan dan oksidasi.
7. **total sulfur dioxide** (Total Sulfur Dioksida): Jumlah total sulfur dioksida dalam anggur, baik yang dalam bentuk bebas maupun yang terikat dengan senyawa lain.
8. **density** (Massa Jenis): Massa jenis anggur, yang merupakan massa anggur per unit volume.
9. **pH**: Ukuran tingkat keasaman atau kebasaan anggur. Skala pH berkisar dari 0 (sangat asam) hingga 14 (sangat basa), dengan 7 sebagai netral.
10. **sulphates** (Sulfat): Jumlah sulfat (biasanya kalium sulfat) dalam anggur.
11. **alcohol** (Alkohol): Persentase kandungan alkohol etil berdasarkan volume (% vol) dalam anggur.
12. **quality** (Kualitas): Skor kualitas anggur yang diberikan oleh panelis ahli. Dalam dataset ini, skornya berkisar dari 3 hingga 8, di mana skor yang lebih tinggi menunjukkan kualitas yang lebih baik.

```python
#Heatmap Korelasi

plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap Korelasi Antar Fitur')
plt.show()
```

Penjelasan korelasi warna yang ada pada Heatmap
* Warna merah → korelasi positif (misal: alcohol ↔ quality)
* Warna biru → korelasi negatif (misal: volatile acidity ↔ quality)
* Warna putih/abu → hampir tidak ada korelasi
* Semakin gelap warnanya (merah atau biru), semakin kuat korelasinya.

Beberapa korelasi antarfitur yang penting:

* fixed acidity vs citric acid = +0.67: cukup tinggi → bisa berdampak pada multikolinearitas jika digunakan dalam model regresi.
* free sulfur dioxide vs total sulfur dioxide = +0.67: sangat wajar karena satu adalah bagian dari yang lain.
* density vs residual sugar = +0.36: logis karena gula meningkatkan massa jenis.

Kesimpulan pada heatmap korelasi yaitu:

* Fitur alcohol, sulphates, dan citric acid cenderung berkontribusi positif terhadap kualitas anggur merah.
* Fitur volatile acidity, density, dan chlorides menunjukkan pengaruh negatif terhadap kualitas.
* Fitur seperti residual sugar, free sulfur dioxide, dan pH memiliki pengaruh yang sangat lemah atau tidak signifikan terhadap kualitas.

# **5. Data Preprocessing**

```python
# Split fitur X dan target Y

X = df.drop('quality', axis=1)
y = df['quality']
```

```python
print("Bentuk Fitur (X):", X.shape)
print("Bentuk Target (y):", y.shape)
print("\nNama kolom fitur:", X.columns.tolist())
```

```python
# Melakukan scalling
scaler = StandardScaler()  # Atau RobustScaler() jika data mengandung outlier

X_scaled_array = scaler.fit_transform(X)

# Mengubah hasil scaling ke DataFrame
X_scaled = pd.DataFrame(X_scaled_array, columns=X.columns)
```

```python
print("Lima baris pertama fitur setelah scaling:")
print(X_scaled.head())
```

# 6. **Menyimpan Data**

```python
# Gabungkan kembali fitur yang sudah di-scale dengan target
df_processed = pd.concat([X_scaled, y.reset_index(drop=True)], axis=1)

# Simpan ke file CSV
df_processed.to_csv('wine_red_preprocessed.csv', index=False)

print("Dataset berhasil disimpan sebagai 'wine_red_preprocessed.csv'")
```
